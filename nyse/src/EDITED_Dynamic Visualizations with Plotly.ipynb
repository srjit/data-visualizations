```python
import plotly.plotly as py
import plotly.graph_objs as go
import random
import numpy as np

l= []
y=[]
# Setting colors for plot.
N= 53
c= ['hsl('+str(h)+',50%'+',50%)' for h in np.linspace(0, 360, N)]

for i in range(int(N)):
    trace0= go.Scatter(
        x= df.curb_w8,
        y= df.horsepower,
        mode= 'markers',
        marker= dict(size= 14,
                    line= dict(width=1),
                    color= c[i],
                    opacity= 0.3
                   ),
        text= df.make)
    l.append(trace0);

layout= go.Layout(
    title= 'Horsepower vs. Weight by car model',
    hovermode= 'closest',
    xaxis= dict(
        title= 'Curb Weight',
        ticklen= 5,
        zeroline= False,
        gridwidth= 2,
    ),
    yaxis=dict(
        title= 'Horsepower',
        ticklen= 5,
        gridwidth= 2,
    ),
    showlegend= False
)
fig= go.Figure(data=l, layout=layout)
py.iplot(fig)
```

This visualization represents a scatterplot of weight vs horsepower of the car models in the dataset. I used a scatterplot representation to depict any correlation that these two variables might have. It can be observed that they have a direct correlation and roughly, more the weight of the car, more is the horsepower value. This gives a general idea of acceleration performance of the car with a few outliers. For example, Porsche apparently has less weight and a horsepower value close to all the high performing cars. The color chosen to depict the scatterplot is Crimson red since it it clearly visible with the white background and provides easy readability of values. The car model variable 'make' is encoded in the scatterplot points.

The interactivity shown in this visualization is when you hover over each point with mouse, it displays the x and y coordinates at the point i.e. the curb weight and horsepower value is displayed, along with the name of the car model. This is important because in a scatter plot a lot if the points are partially/completely overlapped. In that case, to check the value at a particular point is not easy. Hence, this interactive visualization provides the car model name, its name and horsepower values at any given point in the graph.



```python
import plotly.plotly as py
import plotly.graph_objs as go

x = df.fuel_type

trace0 = go.Box(
    y=df.city_mpg,
    x=x,
    name='city',
    marker=dict(
        color='#3D9970'
    )
)
trace1 = go.Box(
    y=df.highway_mpg,
    x=x,
    name='highway',
    marker=dict(
        color='#FF4136'
    )
)

data = [trace0, trace1]
layout = go.Layout(
    title='City vs Highway MPG for gas and diesel cars',
    yaxis=dict(
        title='Miles Per Gallon',
        zeroline = False
        ),
    boxmode = 'group'
)

fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```

This visualization shows the boxplot representation the performance of gas and diesel cars in city and highway. I used a boxplot to represent this data because boxplots are a good way to convey the description of the data like median, quartiles, minimum and maximum data values and outliers. It can be observed that the gas cars display a slightly lower performance compared to diesel cards overall even though there are a few outliers which show that in exceptional models, the performance of a gas car can go as high as 54MPG which is higher than the maximum of diesel cars. Also, generally the car performance reduces in city as compared to on the highway.
Boxplots are extremely helpful in graphically displaying a variable's spread to get an overall sense. It provides some hints towards data's symmetry/non symmetry. Comparison is easy with boxplots and the biggest advantage is that they show outliers.

The interactivity shown in this visualization is when you hover over each box plot with mouse, it displays all the boxplot parameter values such as max, upper fence, upper quartile( q3), median, lower quartile(q1), min. It also shows the values of the outlier points when you mouse over them. This feature allows clear readability of the values of parameters since it is difficult to map the representation to x and y axes and figure out the exact values with naked eye. It also displays whether each boxplot belongs to city or highway when you mouse over them.
